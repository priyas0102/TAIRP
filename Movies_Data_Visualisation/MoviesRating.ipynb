```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
moviesdf = pd.read_csv('movies_metadata.csv',dtype='unicode')
moviesdf.head()
```

```python
moviesdf.info()
```

#### All columns have type object because they most likely contain null values and is loaded in that form as guessing dtypes of each column is very memory demanding, especially when your dataset is large as Pandas tries to determine what dtype to set by analyzing the data in each column. 
#### The total number of rows is 45466 while almost none of the rows have non-null values as 45466. 
#### So we can see that a lot of cleaning is to be done on the data


```python
moviesdf['revenue']=pd.to_numeric(moviesdf['revenue'], errors='coerce')
moviesdf['budget']=pd.to_numeric(moviesdf['budget'], errors='coerce')
moviesdf['popularity']=pd.to_numeric(moviesdf['popularity'], errors='coerce')
moviesdf['vote_average']=pd.to_numeric(moviesdf['vote_average'], errors='coerce')
moviesdf['vote_count']=pd.to_numeric(moviesdf['vote_count'], errors='coerce')
moviesdf['runtime']=pd.to_numeric(moviesdf['runtime'], errors='coerce')
moviesdf['id']=pd.to_numeric(moviesdf['id'], errors='coerce')
```

```python
moviesdf.info()
```

```python
moviesdf.describe()
```

#### We have the desired columns in numeric values, there are a lot of issues with them.
#### The number of outliers present in these columns are significant, so we remove those outliers.



```python
#Cleaning the data by removing outliers
moviesdf.drop(moviesdf[moviesdf['budget']>45000000].index,inplace=True)
moviesdf.drop(moviesdf[moviesdf['budget']<5500000].index,inplace=True)
moviesdf.drop(moviesdf[moviesdf['popularity']<3].index,inplace=True)
moviesdf.drop(moviesdf[moviesdf['popularity']>30].index,inplace=True)
moviesdf.drop(moviesdf[moviesdf['revenue']<10000000].index,inplace=True)
moviesdf.drop(moviesdf[moviesdf['revenue']>90000000].index,inplace=True)
moviesdf.drop(moviesdf[moviesdf['vote_count']<50].index,inplace=True)
moviesdf.drop(moviesdf[moviesdf['vote_count']>800].index,inplace=True)
moviesdf.drop(moviesdf[moviesdf['vote_average']<1].index,inplace=True)
```

#### We can now see that the outliers have been removed, and the variation among the values of the columns like budget, popularity, revenue, vote_count and vote_average have been minimized.

```python
moviesdf.describe()
```

#### Now let's deal with the null values. We can first calculate the percentage of null values in each column.

```python
null_vals_percent=moviesdf.isna().sum()*100/moviesdf.shape[0]
null_vals_percent
```

```python
moviesdf.drop(['belongs_to_collection','genres','homepage','imdb_id','poster_path','overview','production_companies',
                  'production_countries','tagline','original_title','video','id'],axis=1,inplace=True)
```

Now we can drop all the NaN values from the remaining columns



```python
moviesdf.dropna(inplace=True)
moviesdf.shape
```

## Exploratory Analysis and Visualization
#### Now that we have completed the data cleaning and our data is ready to be used, we can use matplotlob and seaborn libraries to perform data visualization. 
#### We can make interesting graphs for various columns of the dataframe and make necessary inferences from it.

 ###### import  matplotlib.pyplot and seaborn.

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 10
matplotlib.rcParams['figure.figsize'] = (6, 6)
matplotlib.rcParams['figure.facecolor'] = '#00000000'

```

#### We can see clearly the basic stats for the revenue generated by movies using barplot in seaborn.



```python
sns.barplot(x=moviesdf.describe()['revenue'].index,y=moviesdf.describe()['revenue'],data = moviesdf)
```

```python
sns.barplot(x=moviesdf.describe()['budget'].index,y=moviesdf.describe()['budget'],data = moviesdf)
```

#### We can plot a histogram, showing the variations across the popularity of the movies.

#### This is a scatterplot having budget on x-axis and revenue on y-axis, and popularity as hue
